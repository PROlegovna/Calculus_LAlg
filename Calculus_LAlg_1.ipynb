```python
5. –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ, —Å—á–∏—Ç–∞–µ—Ç—Å—è –ª–∏ –ª–∏–Ω–µ–π–Ω—ã–º –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º:
–∞) —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –≤—Å–µ—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ —Ç—Ä—ë—Ö–º–µ—Ä–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, —É –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –æ–¥–Ω–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ä–∞–≤–Ω–∞ –Ω—É–ª—é;
–±) –≤—Å–µ –≤–µ–∫—Ç–æ—Ä—ã, —Å—á–∏—Ç–∞—é—â–∏–µ—Å—è –ª–∏–Ω–µ–π–Ω—ã–º–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è–º–∏ –¥–∞–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ {ùë¢1,ùë¢2‚Ä¶,ùë¢ùëõ}.
```

```python
# –∞) —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –≤—Å–µ—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ —Ç—Ä–µ—Ö–º–µ—Ä–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, —É –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –æ–¥–Ω–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ä–∞–≤–Ω–∞ –Ω—É–ª—é.
# –ß—Ç–æ–±—ã –¥–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ, —É –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –æ–¥–Ω–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ä–∞–≤–Ω–∞ –Ω—É–ª—é,
# —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω—ã–º –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ —ç—Ç–æ –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω—ã–º 


# –í–æ—Å–µ–º—å —É—Å–ª–æ–≤–∏–π –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞:

# –û–ø–µ—Ä–∞—Ü–∏—è —Å–ª–æ–∂–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–æ–≤ –¥–æ–ª–∂–Ω–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—Ç—å —Å–ª–µ–¥—É—é—â–∏–º —É—Å–ª–æ–≤–∏—è–º:
# 1) –∫–æ–º–º—É—Ç–∞—Ç–∏–≤–Ω–æ—Å—Ç—å: u + v = v + u ;
# 2) –∞—Å—Å–æ—Ü–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å: (u + v) + w = u + (v + w) ;
# 3) —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ —Ç–∞–∫–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ 0 ‚àà L1 , —á—Ç–æ u + 0 = u –¥–ª—è –ª—é–±–æ–≥–æ u ;
# 4) —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –¥–ª—è –ª—é–±–æ–≥–æ u‚ààV —Ç–∞–∫–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ ‚àíu ‚àà L1 , —á—Ç–æ u + (‚àíu) = 0 .

# –û–ø–µ—Ä–∞—Ü–∏—è —É–º–Ω–æ–∂–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –Ω–∞ —Å–∫–∞–ª—è—Ä –¥–æ–ª–∂–Ω–∞ –ø–æ–¥—á–∏–Ω—è—Ç—å—Å—è —Å–ª–µ–¥—É—é—â–∏–º —É—Å–ª–æ–≤–∏—è–º:
# 5) 1 ‚ãÖ u = u –¥–ª—è –ª—é–±–æ–≥–æ u ‚àà L1 ;
# 6) –∞—Å—Å–æ—Ü–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü–∏–∏ —É–º–Ω–æ–∂–µ–Ω–∏—è –Ω–∞ —Å–∫–∞–ª—è—Ä: Œª(Œº ‚ãÖ u) = (Œª ‚ãÖ Œº)u ;
# 7) –¥–∏—Å—Ç—Ä–∏–±—É—Ç–∏–≤–Ω–æ—Å—Ç—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å–ª–æ–∂–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–æ–≤: Œª(u + v) = Œªu + Œªv ;
# 8) –¥–∏—Å—Ç—Ä–∏–±—É—Ç–∏–≤–Ω–æ—Å—Ç—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å–ª–æ–∂–µ–Ω–∏—è —Å–∫–∞–ª—è—Ä–æ–≤: (Œª + Œº)u = Œªu + Œºu .
```

```python
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')
```

```python
def comparison(one, another):
    display(simplify(one))
    print('==')
    display(simplify(another), simplify(one) == simplify(another))
```

```python
u = Array([0, 0, x])
u
```

```python
v = Array([x, 0, y])
v
```

```python
w = Array([0, y, z])
w
```

```python
# 1) –∫–æ–º–º—É—Ç–∞—Ç–∏–≤–Ω–æ—Å—Ç—å
comparison(u + v, v + u)
```

```python
# 2) –∞—Å—Å–æ—Ü–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å
comparison((u + v) + w, u + (v + w))
```

```python
# –∞) –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –¥–ª–∏–Ω –≤–µ–∫—Ç–æ—Ä–æ–≤;
x1 = Symbol('x1')
x2 = Symbol('x2')
x3 = Symbol('x3')
y1 = Symbol('y1')
y2 = Symbol('y2')
y3 = Symbol('y3')
z1 = Symbol('z1')
z2 = Symbol('z2')
z3 = Symbol('z3')
```

```python
np.random.seed(45)
```

```python
x = Matrix([x1, x2, x3])
x
```

```python
x_l = sqrt(x.dot(x))
x_l
```

```python
y = Matrix([y1, y2, y3])
y
```

```python
y_l = sqrt(y.dot(y))
y_l
```

```python
# –ø—Ä–æ–≤–µ—Ä–∏–º, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –ª–∏ —É—Å–ª–æ–≤–∏—è —É—Å–ª–æ–≤–∏—è–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –µ–≤–∫–ª–∏–¥–æ–≤–∞ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞
# 1. (ùë•,ùë¶)=(ùë¶,ùë•)
comparison(x_l*y_l, y_l*x_l)
```

```python
x_Œª_l*y_l
```

```python
# 2. (Œªùë•,ùë¶)=Œª(ùë•,ùë¶)
display(simplify(x_Œª_l*y_l), Œª * x_l*y_l)
```

```python
z = Matrix([z1, z2, z3])
z
```

```python
z_l = sqrt(z.dot(z))
z_l
```

```python
x_z_l = sqrt((x+z).dot(x+z))
x_z_l
```

```python
# 3. (ùë•1+ùë•2,ùë¶)=(ùë•1,ùë¶)+(ùë•2,ùë¶)
comparison(x_z_l*y_l, simplify(x_l * y_l + z_l*y_l))
```
